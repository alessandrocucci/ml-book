```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

def funzione_retta(x):
    return 3*x+1

X = np.arange(10)
y = funzione_retta(X)

plt.plot(X, y, label='y=3x+1')
plt.plot(X, X**2, label='y=x^2')
plt.legend()
```

```python
df = df.groupby(df.eta).mean().reset_index()
```

```python
x = np.linspace(0, 10, 200)
y = np.cumsum(np.random.randn(200, 3), axis=0)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

fig = plt.gcf()
plt.figure(figsize=(10,10))
plt.plot(x, y)
etichette = ['storico{}'.format(i) for i in range(1,4)]
plt.legend(etichette, loc='upper left')
fig = plt.gcf()
```

```python
import seaborn as sns
sns.set()

fig = plt.gcf()
plt.figure(figsize=(10,10))
plt.plot(x, y)
etichette = ['storico{}'.format(i) for i in range(1,4)]
plt.legend(etichette, loc='upper left')
fig = plt.gcf()
```
